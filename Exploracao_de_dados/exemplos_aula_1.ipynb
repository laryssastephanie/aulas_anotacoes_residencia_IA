```python
%matplotlib inline
```

```python
  import matplotlib.pyplot as plt
  import numpy as np
```

```python
x = np.linspace(0, 10, 100)
```

```python
x.shape # mostra a forma
```

```python
plt.plot(x, np.sin(x), '-k') # x e valor de y (seno de x)
# se add ';' ele omite o objeto
plt.plot(x, np.cos(x), '--r'); # linha tracejada
```

Aspectos Visuais


```python
plt.plot(x, np.sin(x), color="blue", linestyle="dotted");
# a cor pode ser em hexadecimal
# linha tracejada = dashed
# linha pontilhada = dotted
```

Delimitando os eixos

```python
plt.plot(x, np.sin(x), "--r");
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

Adicionando rótulos

```python
plt.plot(x, np.sin(x))
plt.title("Plot do Seno")
plt.xlabel("x")
plt.ylabel("Sin(x)");
```

## Scatter Plots (Dispersão)

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
plt.plot(x, y, 'o', color='black');
```

```python
plt.plot(x, y, '-og'); # linha continua, com pontos, na cor verde
```

### Utilizando o `plt.scatter()`

```python
plt.scatter(x, y, marker='o');
# estilos: o, v, ',', *
```

### Bubble Plot

```python
rng = np.random.RandomState(0)
```

```python
x = rng.randn(100)
```

```python
y = rng.randn(100)
```

```python
z = rng.randn(100)
```

```python
tamanhos = 1000 * rng.rand(100)
```

```python
plt.scatter(x, y, c=z, s=tamanhos, alpha=.3, cmap='rainbow');
# s = size
# alpha = transparencia para visualizar melhor
# cmap = colormap = esquema de cores = verificar documentação
plt.colorbar();
```

### Barras de Erro

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50) # adicionar um ruído
plt.errorbar(x, y, yerr=dy, fmt="o", color="black", ecolor="lightgray"); # colocar a cor do erro diferente para melhorar a visualização
```
